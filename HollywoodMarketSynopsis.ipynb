```python
x = list(range(0,(2020-1995)))
y = list(AnnualTicketSales['TOTAL BOX OFFICE'] )
y.reverse()
y.pop()
y.pop()
slope, intercept, r, p, std_err = stats.linregress(x, y)
x1 = list(range(0,(2022-1995)))
y1= [slope * x + intercept for x in x1 ]
y1.reverse()
AnnualTicketSales['TOTAL BOX OFFICE WITHOUT COVID'] = y1
AnnualTicketSales["Diff"] = AnnualTicketSales['TOTAL BOX OFFICE WITHOUT COVID']-AnnualTicketSales['TOTAL BOX OFFICE']
```

```python
px.line(AnnualTicketSales
       ,x = 'YEAR'
       ,y = ["TOTAL BOX OFFICE","TOTAL BOX OFFICE WITHOUT COVID"]
       ,labels = {'YEAR' :"Years", "value": "Total Sale"}
       ,title = 'TOTAL BOX OFFICE vs TOTAL BOX OFFICE WITHOUT COVID')
```

```python
px.bar(AnnualTicketSales
       ,x = 'YEAR'
       ,y = "Diff"
       ,labels = {'YEAR' :"Year", "Diff": "Financial Loss"}
       ,title = 'Financial Loss (just last two years are important)'
       ,barmode='group')
```

```python
AnnualTicketSales["Percentage of Financial Loss"] = (AnnualTicketSales["TOTAL BOX OFFICE WITHOUT COVID"]-AnnualTicketSales["TOTAL BOX OFFICE"])/AnnualTicketSales["TOTAL BOX OFFICE WITHOUT COVID"]*100
px.bar(AnnualTicketSales
       , x = 'YEAR'
       , y = "Percentage of Financial Loss"
       ,labels = {'YEAR' :"Year", "Percentage of Financial Loss": "Percentage of Financial Loss %"}
       ,title = 'Financial Loss % (just last two years are important) ')
```
